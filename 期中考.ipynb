```
install.packages("SportsAnalytics") #安裝SportsAnalytics
library(SportsAnalytics)       #載入SportsAnalytics
NBA1516 <- fetch_NBAPlayerStatistics("15-16") #定義NBA1516為資料檔
```

```
#顯示NBA1516
str(NBA1516)
```

```
#列出TotalPoints最高的前20個Team
NBA1516Order <- NBA1516[order(NBA1516$TotalPoints,decreasing = T),]
NBA1516Order[1:20,"Team"]
```

```
#列出Position為PG的項目
subset(NBA1516,Position=="PG")
```

```
#列出Name裡面有"J"的項目
NBA1516[grepl("J",NBA1516$Name),]
```

```
#安裝+載入ggplot
install.packages("ggplot2")
library(ggplot2)
```

```
#畫出各Position的TotalRebounds散布圖
#將Position著色
qplot(TotalRebounds , Position , data = NBA1516 , color = Position)
```

```
#畫出TotalRebounds在Position的直方圖
#將Position著色
qplot(TotalRebounds , data = NBA1516 ,fill = Position)
```

```
#畫出X=TotalRebounds與Y=Steals的分群畫圖(橫向)
qplot(TotalRebounds , Steals  , data = NBA1516 ,facets = Position~.)
#畫出X=TotalRebounds與Y=Steals的分群畫圖(直向)
qplot(TotalRebounds , Steals  , data = NBA1516 ,facets = .~Position)
```

```
#畫出X=TotalRebounds並依照Position來分組的分群畫圖，color = Position是將Position上色
qplot(TotalRebounds , data = NBA1516 , facets = Position~. , binWidth = 10 , color = Position)
```
